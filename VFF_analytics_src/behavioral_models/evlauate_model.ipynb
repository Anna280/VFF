## Calculation of log-loss

```python

# === Set temperature value here ===
TEMPERATURE = 2 # Try values like 1.0, 2.0, 3.0, etc.

# Load your CSV file
df = pd.read_csv("/Users/annadaugaard/Desktop/VFF/VFF_analytics_src/results/simple/player_0_best_outputs.csv")
df = df.drop([1, 3, 6, 9]) # for player 0
#df = df.drop([1,3]) # for player 1

y_true = []
y_pred = []

# Process each row
for _, row in df.iterrows():
    scores = np.array(eval(row['scores']))         
    scaled_scores = scores / TEMPERATURE               
    probs = softmax(scaled_scores)                   
    y_true.append(int(row['chosen_target_index']))      
    y_pred.append(probs)                                

# Convert to numpy arrays
y_true = np.array(y_true)
y_pred = np.array(y_pred)

# Ensure class labels are properly specified (0 to number of classes - 1)
n_classes = y_pred.shape[1]
labels = np.arange(n_classes)

# Compute multiclass log-loss
loss = log_loss(y_true, y_pred, labels=labels)
print(f"Multiclass Log-Loss (T={TEMPERATURE}): {loss:.4f}")
```
